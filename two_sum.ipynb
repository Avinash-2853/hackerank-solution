```python
from typing import List

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
       for i in nums:
        for j in nums:
            if (nums.index(i) != nums.index(j)) & (i+j == target):
                return [nums.index(i), nums.index(j)]
```

```python
arr = [2, 7, 11, 15]
target = 9

result = Solution().twoSum(arr, target)
result
```
