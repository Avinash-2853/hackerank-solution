```python
s = "AABCAAADA"
k = 3

def make_unique(chunk):
    unique = []
    for char in chunk:
        if char not in unique:
            unique.append(char)
    return "".join(unique)

for i in range(0, len(s), k):
    chunk = s[i:i+k]
    unique_chunk = make_unique(chunk)
    print(unique_chunk)
```
